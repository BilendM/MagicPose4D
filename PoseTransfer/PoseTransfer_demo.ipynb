```python
import numpy as np
import random
from collections import deque

def rearrange_joints(joints, edges, root=-55):
    num_joints = len(joints)
    neighbors = {i: [] for i in range(num_joints)}
    for edge in edges:
        neighbors[edge[0]].append(edge[1])
        neighbors[edge[1]].append(edge[0])

    # 选择一个端点作为根节点
    if root == -55:
        root = random.choice([i for i in range(num_joints) if len(neighbors[i]) <= 2])
    visited = set()
    queue = deque([root])

    new_order = []
    parent_indices = [-1] * num_joints

    # BFS遍历来重排关节顺序
    while queue:
        node = queue.popleft()
        visited.add(node)
        new_order.append(node)

        for neighbor in neighbors[node]:
            if neighbor not in visited:
                parent_indices[neighbor] = node
                queue.append(neighbor)

    # 重新排列joints和parent_indices
    new_joints = np.array([joints[i] for i in new_order])
    new_parent_indices = [new_order.index(parent_indices[i]) if parent_indices[i] != -1 else -1 for i in new_order]

    return new_joints, new_parent_indices, root


def convert_fk_to_original_format(points, parent_indices):
    edges = []

    for idx, parent_idx in enumerate(parent_indices):
        if parent_idx is not None and parent_idx != -1:
            edges.append([parent_idx, idx])

    return np.array(points), np.array(edges)

import numpy as np

def calculate_joint_directions(points, parent_indices):
    directions = []
    for i, parent_index in enumerate(parent_indices):
        if parent_index is None or parent_index == -1:
            directions.append(np.array([0, 0, 0]))  # 根节点没有方向
        else:
            vector = points[i] - points[parent_index]
            direction = vector / np.linalg.norm(vector) if np.linalg.norm(vector) != 0 else np.array([0, 0, 0])
            directions.append(direction)
    return directions

def apply_directions_to_skeleton(initial_points, directions, parent_indices):
    new_positions = [initial_points[0]]  # 根节点位置保持不变
    for i in range(1, len(initial_points)):
        direction = directions[i]
        parent_index = parent_indices[i]
        if parent_index == -1:  # 对于根节点
            continue
        # 计算连杆长度
        link_length = np.linalg.norm(initial_points[i] - initial_points[parent_index])
        # 更新新的关节位置
        new_position = new_positions[parent_index] + direction * link_length
        new_positions.append(new_position)
    return np.array(new_positions)


import numpy as np
from scipy.spatial.transform import Rotation as R

def compute_transformations(s1, s2, parent_index):
    """
    Compute the transformation matrices needed to transform s1 into s2.

    :param s1: N x 3 matrix representing the original skeleton.
    :param s2: N x 3 matrix representing the transformed skeleton.
    :param parent_index: List of parent indices for each joint.
    :return: List containing the transformation matrices for each joint.
    """

    # Ensure s1 and s2 are float type
    s1 = s1.astype(float)
    s2 = s2.astype(float)

    N = s1.shape[0]
    transformations = []
    transformations_e = []

    for i in range(N):
        # Compute translation
        translation = np.zeros(3) if parent_index[i] == -1 else s2[parent_index[i]] - s1[parent_index[i]]

        # Compute rotation
        if parent_index[i] != -1:
            v1 = s1[i] - s1[parent_index[i]]
            v2 = s2[i] - s2[parent_index[i]]
            v1 /= np.linalg.norm(v1)
            v2 /= np.linalg.norm(v2)

            cross_prod = np.cross(v1, v2)
            dot_prod = np.dot(v1, v2)
            if np.linalg.norm(cross_prod) != 0:
                rotation_axis = cross_prod / np.linalg.norm(cross_prod)
                rotation_angle = np.arccos(dot_prod)
                rotation = R.from_rotvec(rotation_angle * rotation_axis)
            else:
                rotation = R.from_rotvec(np.zeros(3))
        else:
            rotation = R.from_rotvec(np.zeros(3))

        # Create transformation matrix
        translation_to_origin = np.eye(4)
        translation_to_origin[:3, 3] = -s1[parent_index[i]]

        # 创建平移矩阵以将旋转点从原点移回
        translation_back = np.eye(4)
        translation_back[:3, 3] = s2[parent_index[i]]
        rotation_matrix = rotation.as_matrix()

        # 组合变换矩阵
        transformation_matrix = translation_back @ np.block([[rotation_matrix, np.zeros((3,1))], [0,0,0,1]]) @ translation_to_origin
        transformations.append(transformation_matrix)

    return np.array(transformations)




def apply_transformations(transformations, s1, parent_index):
    """
    Apply the computed transformations to s1.

    :param transformations: List of transformation matrices for each joint.
    :param s1: N x 3 matrix representing the original skeleton.
    :param parent_index: List of parent indices for each joint.
    :return: N x 3 matrix representing the transformed skeleton.
    """
    N = s1.shape[0]
    s2 = np.zeros_like(s1)

    for i in range(N):
        if parent_index[i] == -1:
            s2[i] = s1[i]
        else:
            joint_pos = transformations[i] @ np.append(s1[i], 1)
            s2[i] = joint_pos[:3]

    return s2


def blend_skinning(transformations, parent_index, mesh1, skin_weights):
    """
    Apply Linear Blend Skinning to transform mesh1 to a new pose.

    :param transformations: List of transformation matrices for each joint.
    :param parent_index: List of parent indices for each joint.
    :param mesh1: N x 3 matrix representing the original mesh vertices.
    :param skin_weights: N x M matrix of skin weights, where N is the number of vertices,
                         and M is the number of joints.
    :return: N x 3 matrix representing the transformed mesh vertices.
    """
    N = mesh1.shape[0]
    M = len(transformations)
    mesh2 = np.zeros_like(mesh1)

    for i in range(N):
        vertex_transform = np.zeros((4, 4))
        for j in range(M):
            # Accumulate weighted transformation for each bone
            weight = skin_weights[i, j]
            bone_transform = transformations[j] if parent_index[j] != -1 else np.eye(4)
            vertex_transform += weight * bone_transform

        # Apply the accumulated transformation to the vertex
        transformed_vertex = vertex_transform @ np.append(mesh1[i], 1)
        mesh2[i] = transformed_vertex[:3]

    return mesh2

def compute_link_lengths(p, parent_indices):
    """
    Compute the length of the link between each point and its parent.

    :param p: An array of N 3D points (N x 3 matrix).
    :param parent_indices: An array of parent indices for each point.
    :return: An array of link lengths.
    """
    N = p.shape[0]
    link_lengths = np.zeros(N)

    for i in range(N):
        if parent_indices[i] != -1:
            link_lengths[i] = np.linalg.norm(p[i] - p[parent_indices[i]])

    return link_lengths

def modify_obj_vertex_x(obj_file_path):
    """
    Modify the vertex coordinates in an OBJ file from (x, y, z) to (x, z, y).

    :param obj_file_path: Path to the OBJ file.
    """
    modified_lines = []

    with open(obj_file_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            if line.startswith('v '):  # Vertex definition
                parts = line.split()
                if len(parts) == 4:
                    x, y, z = parts[1:4]
                    modified_line = f"v {x} {z} {y}\n"
                    modified_lines.append(modified_line)
                else:
                    modified_lines.append(line)
            else:
                modified_lines.append(line)

    with open(obj_file_path, 'w') as file:
        file.writelines(modified_lines)
        
import vtk
import pynocchio as pyn
from pynocchio import skeletons
import numpy as np
import trimesh as tm
import open3d as o3d

from pynocchio import auto_rig, Mesh, Vector3, Transform, VectorTransform, Points, Quaternion


import numpy as np

def write_skeleton_file(joints, parent_joints, filename):
    """
    Writes a skeleton definition to a file.
    
    Parameters:
    - joints: A numpy array of shape (N, 3) representing joint coordinates.
    - parent_joints: A numpy array of shape (N, 1) representing indices of parent joints.
    - filename: The name of the file to which the skeleton will be written.
    """
    with open(filename, 'w') as file:
        for i, (x, y, z) in enumerate(joints):
            # Parent index of -1 indicates no parent, so we write an empty string
            parent = parent_joints[i][0]
            parent_str = str(parent) if parent != -1 else ""
            joint_line = f"joint{i} {x} {y} {z} {parent_str}\n"
            file.write(joint_line)
```

```python
import numpy as np
import pynocchio as pn

def matrix_to_transforms(matrices):
    vector_transforms = []
    for mat in matrices:
        # Extract rotation matrix and translation vector
        rotation_mat = mat[:3, :3]
        trans_vec = mat[:3, 3]
        
        # Calculate rotation axis and angle from rotation_mat
        # Simplified example, real calculation would be more complex
        axis, angle = rotation_matrix_to_axis_angle(rotation_mat)  # Implement this function based on earlier guidance
        # print(rotation_mat)
        # print(axis, angle)
        # print(kkk)
        # Create Quaternion, assuming a function or method to convert axis-angle to quaternion
        
        quaternion = pn.Quaternion(Vector3(axis[0], axis[1], axis[2]), angle)
        
        # Create translation Vector3
        translation = pn.Vector3(*trans_vec)
        
        # Create Transform object
        transform = pn.Transform(quaternion, 1.0, translation)
        
        vector_transforms.append(transform)
    
    # Initialize VectorTransform with the list of Transform objects
    return pn.VectorTransform(vector_transforms)


def rotation_matrix_to_axis_angle(matrix):
    # Ensure the matrix is a numpy array
    matrix = np.asarray(matrix)
    # Calculate the angle
    angle = np.arccos((np.trace(matrix) - 1) / 2.0)
    # Calculate the rotation axis
    rx = matrix[2, 1] - matrix[1, 2]
    ry = matrix[0, 2] - matrix[2, 0]
    rz = matrix[1, 0] - matrix[0, 1]
    norm = np.sqrt(rx**2 + ry**2 + rz**2)
    if norm == 0:
        return np.array([1, 0, 0]), 0
    
    axis = np.array([rx, ry, rz]) / norm
    return axis, np.degrees(angle)
```

```python
import numpy as np

def compute_edge_centers(joints, edges):
    """
    Compute the center points of edges defined by pairs of joint coordinates.

    Parameters:
    joints (array_like): An N x 3 array where each row represents the coordinates of a joint.
    edges (array_like): An M x 2 array where each row represents a pair of indices into the joints array
                        that defines an edge.

    Returns:
    numpy.ndarray: An M x 3 array where each row is the center point of the corresponding edge.
    """
    # Convert inputs to numpy arrays for easier manipulation
    joints = np.array(joints)
    edges = np.array(edges)
    
    # Initialize an array to hold the center points
    edge_centers = np.zeros((len(edges), 3))
    edge_lengths = np.zeros(len(edges))
    
    # Calculate the center point for each edge
    for i, (start_idx, end_idx) in enumerate(edges):
        start_point = joints[start_idx]
        end_point = joints[end_idx]
        edge_centers[i] = (start_point + end_point) / 2
        edge_lengths[i] = np.linalg.norm(end_point - start_point)
        
    return edge_centers, edge_lengths
```

```python
def swap_yz_trimesh(mesh):
    """
    Swap the y and z coordinates of a trimesh mesh's vertices.

    Parameters:
    mesh (trimesh.Trimesh): The input mesh whose vertices are to be modified.

    Returns:
    trimesh.Trimesh: A new trimesh mesh with y and z coordinates swapped.
    """
    # Swap the y and z coordinates of each vertex
    swapped_vertices = mesh.vertices.copy()
    swapped_vertices[:, [1, 2]] = swapped_vertices[:, [2, 1]]
    swapped_vertices[:, 0] = -swapped_vertices[:, 0]  # Negate x

    # Create a new mesh with the swapped vertices
    new_mesh = tm.Trimesh(vertices=swapped_vertices, faces=mesh.faces.copy(), process=False)
    
    return new_mesh

def swap_yz_points(p):
    p[:, [1, 2]] = p[:, [2, 1]]
    p[:, 0] = -p[:, 0]  # Negate x
    return p

import os

def get_subdirectories(path):
    """
    Returns a list of paths to all subdirectories within a given directory.

    Parameters:
    - path: The path to the directory whose subdirectories you want to list.

    Returns:
    - A list of paths to subdirectories.
    """
    # List all entries in the directory given by "path"
    entries = os.listdir(path)
    
    # Combine the entries with the base path to get absolute paths
    full_paths = [os.path.join(path, entry) for entry in entries]
    
    # Filter out the full paths that are directories
    subdirectory_paths = [entry for entry in full_paths if os.path.isdir(entry)]
    
    return subdirectory_paths
```

```python
import open3d as o3d
import numpy as np
import os

# 定义旋转矩阵函数
def get_rotation_matrix(axis, angle):
    if axis == 'x':
        return np.array([
            [1, 0, 0],
            [0, np.cos(angle), -np.sin(angle)],
            [0, np.sin(angle), np.cos(angle)]
        ])
    elif axis == 'y':
        return np.array([
            [np.cos(angle), 0, np.sin(angle)],
            [0, 1, 0],
            [-np.sin(angle), 0, np.cos(angle)]
        ])
    elif axis == 'z':
        return np.array([
            [np.cos(angle), -np.sin(angle), 0],
            [np.sin(angle), np.cos(angle), 0],
            [0, 0, 1]
        ])

def rotate_obj_files(file_paths, rotation_sequence):
    for file_path in file_paths:
        # 读取 OBJ 文件
        mesh = o3d.io.read_triangle_mesh(file_path)
        
        # 应用旋转序列
        R = np.eye(3)
        for axis, degrees in rotation_sequence:
            angle = np.radians(degrees)
            R = R @ get_rotation_matrix(axis, angle)
        
        # 应用旋转矩阵
        mesh.rotate(R, center=(0, 0, 0))
        
        # 保存 OBJ 文件
        o3d.io.write_triangle_mesh(file_path, mesh)
        
        
def get_subdirectories(path):
    """
    Returns a list of paths to all subdirectories within a given directory.

    Parameters:
    - path: The path to the directory whose subdirectories you want to list.

    Returns:
    - A list of paths to subdirectories.
    """
    # List all entries in the directory given by "path"
    entries = os.listdir(path)
    
    # Combine the entries with the base path to get absolute paths
    full_paths = [os.path.join(path, entry) for entry in entries]
    
    # Filter out the full paths that are directories
    subdirectory_paths = [entry for entry in full_paths if os.path.isdir(entry)]
    
    return subdirectory_paths
```

```python
import trimesh
import numpy as np

def bone_from_skin(obj_path, weights, p, outpth, v_o, i = 0):
    """
    计算给定 OBJ 文件中各部分的加权中心点。

    参数:
    - obj_path: OBJ 文件的路径。
    - weights: 一个 (N, B) 形状的数组，N 是顶点数，B 是骨骼数（部分数），表示每个顶点对应的权重。

    返回:
    - centers: 每个部分的中心点坐标，形状为 (B, 3)。
    """
    # 读取 OBJ 文件
    mesh = trimesh.load(obj_path)
    raw_mesh = o3d.io.read_triangle_mesh(obj_path)

    # 获取顶点坐标
    vertices = np.array(raw_mesh.vertices)  # 形状为 (N, 3)

    # 确保权重矩阵的形状匹配
    if vertices.shape[0] != weights.shape[0]:
        print(vertices.shape, weights.shape)
        raise ValueError("顶点数和权重矩阵的顶点数不匹配。")
        

    # 计算每个部分的加权中心点
    centers = np.zeros((weights.shape[1], 3))  # 初始化中心点数组，形状为 (B, 3)
    for i in range(weights.shape[1]):  # 遍历每个骨骼/部分
        # 计算第 i 个骨骼的加权中心
        total_weight = np.sum(weights[:, i])  # 计算第 i 列的权重和
        if total_weight > 0:
            weighted_positions = weights[:, i][:, np.newaxis] * vertices  # 加权位置
            centers[i] = np.sum(weighted_positions, axis=0) / total_weight  # 加权平均

    indices = np.where(p == 0)[0]
    #print(indices)
    indices -= 1
    #print(centers[indices])
    first_row = centers[indices].mean(0)
    #print(first_row)
    #print(ddd)
    centers2 = np.vstack([first_row, centers])

    joi2, edg2 = convert_fk_to_original_format(centers2, p)

    vis_skl(mesh, raw_mesh, edg2, joi2, f'{outpth}/bones_from_skin{i}')
    
    return centers2


def compute_bone_parents(parent_indices):
    """
    计算骨骼间的连接关系。

    参数:
    - parent_indices: 关节的父节点索引数组，其中第一个元素是-1（根节点）

    返回:
    - bone_parents: 骨骼的父节点索引数组
    """
    # 初始化骨骼的父节点索引数组，长度为 N-1
    N = len(parent_indices)
    bone_parents = np.full(N-1, -1, dtype=int)  # 先填充为-1

    # 遍历每个关节来定义骨骼的父节点
    for i in range(1, N):  # 从 1 开始，跳过根节点
        parent_joint = parent_indices[i]
        if parent_joint == 0:
            # 如果父关节是根关节，该骨骼父节点设置为-1
            bone_parents[i-1] = -1
        else:
            # 父骨骼的索引是父关节的索引减一
            bone_parents[i-1] = parent_joint - 1

    return bone_parents


import pdb
# def ptrans(list_obj1, obj2, outpth):
    
#     mesh_out = outpth
#     pig_out = '/u/haozhang/files/4D data/DeformingThings4D/experiments/pig'
    
#     attach2,v0,p = rigging(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', obj2)
#     p_b = p
#     skin_weights2 = attach2.getAllWeights()

#     visualize_skin_weights(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', skin_weights2, f'/u/haozhang/files/pynocchio/data/{obj2}', pig_out)
#     show_skin(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', skin_weights2.T, f'/u/haozhang/files/pynocchio/data/{obj2}/skin_{obj2}')
#     v2 = bone_from_skin(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', skin_weights2, p_b, pig_out, v0)
#     f_num = count_obj_files(f'/u/haozhang/files/4D data/DeformingThings4D/experiments/{exp_name}/lbs')
    
#     for i in range(f_num):
        
#         skin_weights1 = np.load(f'/u/haozhang/files/4D data/DeformingThings4D/experiments/{exp_name}/skin.npy')
#         joints = np.load(f'/u/haozhang/files/4D data/DeformingThings4D/experiments/{exp_name}/{exp_name}_joints.npy')
#         v1 = bone_from_skin(f'{list_obj1}/{exp_name}_frame_{i}.obj', skin_weights1, p_b, mesh_out, joints, i)
        
#         directions = calculate_joint_directions(v1, p_b)
#         v2_new = apply_directions_to_skeleton(v2, directions, p_b)

#         new_joints2, new_edges2 = convert_fk_to_original_format(v2_new, p_b)
#         vis_skle(new_edges2, new_joints2, f'/u/haozhang/files/pynocchio/data/{obj2}/new_pose_{i}')
#         T = compute_transformations(v2, v2_new, p_b)
        
#         v22 = apply_transformations(T, v2, p_b)
#         j22, e22 = convert_fk_to_original_format(v22, p_b)
#         vis_skle(e22, j22, f'/u/haozhang/files/pynocchio/data/{obj2}/transformed_pose_{i}')
        
#         # print('Reference Object:', obj1)
#         # print('Target Object:', obj2)
#         # print('Joints shape:', v2.shape)
#         # print('Bones shape:', e22.shape)
#         # print('Parent indixes:', p1)
#         # print('Transformation matrix shape:', T.shape)
#         # print('Skinweights shape:', attach2.getAllWeights().shape)
        
#         mesh_path = f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj'
#         mesh = tm.load(mesh_path)
#         vertices_np = blend_skinning(T[1:], p_b[1:], np.array(mesh.vertices), skin_weights2)
#         Tmat = np.load(f'/u/haozhang/files/4D data/DeformingThings4D/experiments/{exp_name}/lbs/T_{exp_name}_{i}.npy')
#         vertices_np += Tmat[0]
#         # print(vertices_np.shape)
#         # pdb.set_trace()
#         # Update the vertices
#         mesh.vertices = vertices_np
#         # print(mesh.vertices)
#         # Save the mesh as an OBJ file
#         mesh.export(f'{mesh_out}/transformaed_mesh_{i}.obj')
        
from sklearn.neighbors import NearestNeighbors

def compute_skin_for_newmesh(m1, E, m2):
    """
    Compute embeddings for m2 based on the closest points in m1.

    Parameters:
    m1 : array
        [n1, 3] shape array representing mesh1 vertices.
    E : array
        [n1, b] shape array representing embeddings for mesh1 (skin weights here).
    m2 : array
        [n2, 3] shape array representing mesh2 vertices.

    Returns:
    E2 : array 
        [n2, b] shape array representing computed embeddings for mesh2 (new skin weights here).
    """

    # Fit the NearestNeighbors model on m1
    nbrs = NearestNeighbors(n_neighbors=1, algorithm='auto').fit(m1)

    # Find the closest points in m1 for each point in m2
    distances, indices = nbrs.kneighbors(m2)

    # Get the embeddings for the closest points to construct E2
    E2 = E[indices.flatten()]

    return E2

        
import pdb
def ptrans(list_obj1, obj2, outpth, exp_name):
    
    mesh_out = outpth
    pig_out = '/u/haozhang/files/4D data/DeformingThings4D/experiments/pig'
    
    if list_obj1.split('/')[-3] == 'experiments':
        obj_type = 'quad'
    else:
        obj_type = 'human'
        
    print(obj_type)
        
    attach2,v0,p = rigging(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', obj2, obj_type)
    p_b = p
    skin_weights2 = attach2.getAllWeights()

    visualize_skin_weights(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', skin_weights2, f'/u/haozhang/files/pynocchio/data/{obj2}', pig_out)
    show_skin(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', skin_weights2.T, f'/u/haozhang/files/pynocchio/data/{obj2}/skin_{obj2}')
    v2 = bone_from_skin(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', skin_weights2, p_b, pig_out, v0)
    f_num = count_obj_files(list_obj1)
    folder = list_obj1.split('/')[-3]
    
    # full_model = tm.load(f'/u/haozhang/files/pynocchio/data/{obj2}/model2.obj')
    # simple_model = tm.load(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj')
    # # print(np.array(simple_model.vertices).shape, skin_weights2.shape, np.array(full_model.vertices).shape)
    # # print(kkk)
    # skin_full = compute_skin_for_newmesh(np.array(simple_model.vertices), skin_weights2, np.array(full_model.vertices))
    
    for i in range(f_num):
        
        full_model = tm.load(f'/u/haozhang/files/pynocchio/data/{obj2}/model2.obj')
        simple_model = tm.load(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj')
        # print(np.array(simple_model.vertices).shape, skin_weights2.shape, np.array(full_model.vertices).shape)
        # print(kkk)
        skin_full = compute_skin_for_newmesh(np.array(simple_model.vertices), skin_weights2, np.array(full_model.vertices))
        
        skin_weights1 = np.load(f'/u/haozhang/files/4D data/DeformingThings4D/{folder}/{exp_name}/skin.npy')
        joints = np.load(f'/u/haozhang/files/4D data/DeformingThings4D/{folder}/{exp_name}/{exp_name}_joints.npy')
        
        # print(skin_weights1.shape, joints.shape)
        # print(kkk)
        v1 = bone_from_skin(f'{list_obj1}/{exp_name}_frame_{i}.obj', skin_weights1, p_b, mesh_out, joints, i)
        
        directions = calculate_joint_directions(v1, p_b)
        v2_new = apply_directions_to_skeleton(v2, directions, p_b)

        new_joints2, new_edges2 = convert_fk_to_original_format(v2_new, p_b)
        vis_skle(new_edges2, new_joints2, f'/u/haozhang/files/pynocchio/data/{obj2}/new_pose_{i}')
        T = compute_transformations(v2, v2_new, p_b)
        
        v22 = apply_transformations(T, v2, p_b)
        j22, e22 = convert_fk_to_original_format(v22, p_b)
        vis_skle(e22, j22, f'/u/haozhang/files/pynocchio/data/{obj2}/transformed_pose_{i}')
        
        # print('Reference Object:', obj1)
        # print('Target Object:', obj2)
        # print('Joints shape:', v2.shape)
        # print('Bones shape:', e22.shape)
        # print('Parent indixes:', p1)
        # print('Transformation matrix shape:', T.shape)
        # print('Skinweights shape:', attach2.getAllWeights().shape)
        
        mesh = simple_model
        vertices_np = blend_skinning(T[1:], p_b[1:], np.array(mesh.vertices), skin_weights2)
        Tmat = np.load(f'/u/haozhang/files/4D data/DeformingThings4D/{folder}/{exp_name}/lbs/T_{exp_name}_{i}.npy')
        vertices_np += Tmat[0]
        # print(vertices_np.shape)
        # pdb.set_trace()
        # Update the vertices
        mesh.vertices = vertices_np
        # print(mesh.vertices)
        # Save the mesh as an OBJ file
        mesh.export(f'{mesh_out}/transformaed_mesh_{i}.obj')
        
        mesh2 = full_model
        vertices_np2 = blend_skinning(T[1:], p_b[1:], np.array(mesh2.vertices), skin_full)
        vertices_np2 += Tmat[0]
        # print(vertices_np.shape)
        # pdb.set_trace()
        # Update the vertices
        mesh2.vertices = vertices_np2
        # print(mesh.vertices)
        # Save the mesh as an OBJ file
        mesh2.export(f'{mesh_out}/colored_transformed_mesh{i}.obj')
        
        


def ptrans2(list_obj1, list_obj2, obj2, outpth, exp_name):
    
    mesh_out = outpth
    pig_out = '/u/haozhang/files/4D data/DeformingThings4D/experiments/pig'
    
    if list_obj1.split('/')[-3] == 'experiments':
        obj_type = 'quad'
    else:
        obj_type = 'human'
    print(obj_type)
    
    attach2,v0,p = rigging(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', obj2, obj_type)
    p_b = p
    skin_weights2 = attach2.getAllWeights()

    visualize_skin_weights(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', skin_weights2, f'/u/haozhang/files/pynocchio/data/{obj2}', pig_out)
    show_skin(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', skin_weights2.T, f'/u/haozhang/files/pynocchio/data/{obj2}/skin_{obj2}')
    v2 = bone_from_skin(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj', skin_weights2, p_b, pig_out, v0)
    
    f_num = count_obj_files(list_obj1)
    folder = list_obj1.split('/')[-3]
        
    
    for i in range(1, f_num):
        if not os.path.exists(f'{list_obj2}/{exp_name}_frame_{i}.obj') or os.path.exists(f'{mesh_out}/colored_transformed_mesh{i}.obj'):
            continue
        
        full_model = tm.load(f'/u/haozhang/files/pynocchio/data/{obj2}/model2.obj')
        simple_model = tm.load(f'/u/haozhang/files/pynocchio/data/{obj2}/sample_{obj2}.obj')
        
        # print(np.array(simple_model.vertices).shape, skin_weights2.shape, np.array(full_model.vertices).shape)
        # print(kkk)
        skin_full = compute_skin_for_newmesh(np.array(simple_model.vertices), skin_weights2, np.array(full_model.vertices))
        
        skin_weights1 = np.load(f'/u/haozhang/files/4D data/DeformingThings4D/{folder}/{exp_name}/skin.npy')
        
        ori_model = tm.load(f'{list_obj2}/{exp_name}_frame_0.obj')
        ref_model = tm.load(f'{list_obj2}/{exp_name}_frame_0_simple.obj')
        skin_ori = compute_skin_for_newmesh(np.array(ref_model.vertices), skin_weights1, np.array(ori_model.vertices))
        
        # print(np.array(ori_model.vertices).shape)
        # print(kkk)
        joints = np.load(f'/u/haozhang/files/4D data/DeformingThings4D/{folder}/{exp_name}/{exp_name}_joints.npy')
        #v1 = bone_from_skin(f'{list_obj1}/{exp_name}_frame_{i}.obj', skin_weights1, p_b, mesh_out, joints, i)
        
        # print(skin_ori.shape, joints.shape)
        
        v1 = bone_from_skin(f'{list_obj2}/{exp_name}_frame_{i}.obj', skin_ori, p_b, mesh_out, joints, i)
        
        directions = calculate_joint_directions(v1, p_b)
        v2_new = apply_directions_to_skeleton(v2, directions, p_b)

        new_joints2, new_edges2 = convert_fk_to_original_format(v2_new, p_b)
        vis_skle(new_edges2, new_joints2, f'/u/haozhang/files/pynocchio/data/{obj2}/new_pose_{i}')
        T = compute_transformations(v2, v2_new, p_b)
        
        v22 = apply_transformations(T, v2, p_b)
        j22, e22 = convert_fk_to_original_format(v22, p_b)
        vis_skle(e22, j22, f'/u/haozhang/files/pynocchio/data/{obj2}/transformed_pose_{i}')
        
        # print('Reference Object:', obj1)
        # print('Target Object:', obj2)
        # print('Joints shape:', v2.shape)
        # print('Bones shape:', e22.shape)
        # print('Parent indixes:', p1)
        # print('Transformation matrix shape:', T.shape)
        # print('Skinweights shape:', attach2.getAllWeights().shape)
        
        mesh = simple_model
        vertices_np = blend_skinning(T[1:], p_b[1:], np.array(mesh.vertices), skin_weights2)
        Tmat = np.load(f'/u/haozhang/files/4D data/DeformingThings4D/{folder}/{exp_name}/lbs/T_{exp_name}_{i}.npy')
        vertices_np += Tmat[0]
        # print(vertices_np.shape)
        # pdb.set_trace()
        # Update the vertices
        mesh.vertices = vertices_np
        # print(mesh.vertices)
        # Save the mesh as an OBJ file
        mesh.export(f'{mesh_out}/transformaed_mesh_{i}.obj')
        
        mesh2 = full_model
        vertices_np2 = blend_skinning(T[1:], p_b[1:], np.array(mesh2.vertices), skin_full)
        vertices_np2 += Tmat[0]
        # print(vertices_np.shape)
        # pdb.set_trace()
        # Update the vertices
        mesh2.vertices = vertices_np2
        # print(mesh.vertices)
        # Save the mesh as an OBJ file
        mesh2.export(f'{mesh_out}/colored_transformed_mesh{i}.obj')

        
        
def count_obj_files(directory):
    """
    计算指定目录中以 '.obj' 结尾的文件数量。

    参数:
    - directory: 要检查的目录路径。
    
    返回:
    - count: 以 '.obj' 结尾的文件数量。
    """
    count = 0
    # 检查目录是否存在
    if not os.path.exists(directory):
        print(f"给定的路径不存在: {directory}")
        return count

    # 遍历目录
    for entry in os.listdir(directory):
        # 构建完整的文件路径
        full_path = os.path.join(directory, entry)
        # 确保是文件且以 .obj 结尾
        if os.path.isfile(full_path) and full_path.endswith('.obj'):
            count += 1

    return count
```

```python
def rigging(mesh_file, name, sk_type='quad', outpth=None):

    raw_mesh = o3d.io.read_triangle_mesh(mesh_file)
    raw_mesh1 = tm.load(mesh_file)

    reader = vtk.vtkOBJReader()
    reader.SetFileName(mesh_file)
    reader.Update()
    poly_data = reader.GetOutput()

    points = pyn.Points()
    for i in range(poly_data.GetPoints().GetNumberOfPoints()):
        p = poly_data.GetPoints().GetPoint(i)
        points.append(pyn.Vector3(p[0], p[1], p[2]))
    triangles = pyn.Indices()

    for i in range(poly_data.GetNumberOfCells()):
        p_ids = vtk.vtkIdList()
        poly_data.GetCellPoints(i, p_ids)
        for p_id in range(p_ids.GetNumberOfIds()):
            triangles.append(p_ids.GetId(p_id))

    if sk_type == 'human':
        skeleton = skeletons.HumanSkeleton()
    elif sk_type == 'quad':
        skeleton = skeletons.HorseSkeleton()
    else:
        skeletons.FileSkeleton(sk_type)
        
    skeleton.scale(.7)
    v = np.array(skeleton.vertices)
    p = np.array(skeleton.parent_indices)
    mesh = pyn.Mesh(points, triangles)
    attach = pyn.auto_rig(skeleton, mesh)

    v2 = np.array(attach.embedding)
    joi2, edg2 = convert_fk_to_original_format(v2, p)
    
    if outpth is None:
        vis_skl(raw_mesh1, raw_mesh, edg2, joi2, f'/u/haozhang/files/pynocchio/data/{name}/{name}')
    else:
        vis_skl(raw_mesh1, raw_mesh, edg2, joi2, f'{outpth}/{name}')
    
    bones, _ = compute_edge_centers(joi2, edg2)
    if outpth is None:
        np.save(f'/u/haozhang/files/pynocchio/data/{name}/{name}_bones.npy', bones)
        vis_skl(raw_mesh1, raw_mesh, edg2, joi2, f'/u/haozhang/files/pynocchio/data/{name}/{name}_bones', bones)
    else:
        np.save(f'{outpth}/{name}_bones.npy', bones)
        vis_skl(raw_mesh1, raw_mesh, edg2, joi2, f'{outpth}/{name}_bones', bones)
        np.save(f'{outpth}/{name}_joints.npy', v2)
        np.save(f'{outpth}/{name}_parent_indices.npy', p)
    
    return attach, v2, p
```

```python
import trimesh
import os
import numpy as np


def rotate_mesh(input_path, output_path):
    # 读取 OBJ 文件
    mesh = trimesh.load_mesh(input_path)
    
    # 定义绕 y 轴顺时针旋转 90 度的旋转矩阵
    rotation_y = np.array([
        [0, 0, 1],
        [0, 1, 0],
        [-1, 0, 0]
    ])
    
    # 定义绕 z 轴顺时针旋转 90 度的旋转矩阵
    rotation_z = np.array([
        [0, -1, 0],
        [1, 0, 0],
        [0, 0, 1]
    ])
    
    # 组合旋转矩阵
    combined_rotation = np.dot(rotation_z, rotation_y)
    
    # 旋转 mesh 的顶点
    mesh.vertices = np.dot(mesh.vertices, combined_rotation.T)
    
    # 保存新的 OBJ 文件
    mesh.export(output_path)
    
    
def keep_largest_watertight_component(input_path, output_path):
    # Check if the input file path exists
    if not os.path.exists(input_path):
        raise FileNotFoundError(f"The input file path {input_path} does not exist")
    
    # Load the OBJ file
    mesh = trimesh.load_mesh(input_path)
    
    # Ensure the mesh is in the correct format
    if not isinstance(mesh, trimesh.Trimesh):
        raise ValueError("The input file does not contain a valid mesh")

    # Split the mesh into connected components
    components = mesh.split(only_watertight=False)
    
    # Find the largest watertight component
    largest_component = None
    largest_volume = 0
    for component in components:
        if component.is_watertight:
            volume = component.volume
            if volume > largest_volume:
                largest_volume = volume
                largest_component = component
    
    # Check if a watertight component was found
    if largest_component is None:
        raise ValueError("No watertight components found")

    # Save the largest watertight component to a new OBJ file
    largest_component.export(output_path)
    
    
import os
root = '/u/haozhang/files/mul_pac/threestudio/outputs/human_meshes'
l1 = os.listdir(root)
print(l1)

import subprocess
for k in l1:
    pth = f'{root}/{k}/save/it800-export/model.obj'
    name = k.split('.')[0]
    # 示例使用
    
    rotate_file = f'{root}/{k}/save/it800-export/model_rotated.obj'
    rotate_mesh(pth, rotate_file)
    out_pth = f'/u/haozhang/files/4D data/DeformingThings4D/experiments/human1/human1_frame_{int(name)}_remesh.obj'
    out_pth2 = f'/u/haozhang/files/4D data/DeformingThings4D/experiments/human1/human1_frame_{int(name)}_simple.obj'
    #if os.path.exists(out_pth2): continue
    print(subprocess.check_output(['/u/haozhang/files/mul_pac/Manifold/build/manifold', rotate_file, out_pth, '-m', '-f', '8000']))
    
    output_file = f'/u/haozhang/files/4D data/DeformingThings4D/experiments/human1/human1_frame_{int(name)}_remesh2.obj'
    keep_largest_watertight_component(out_pth, output_file)
    
    print(subprocess.check_output(['/u/haozhang/files/mul_pac/Manifold/build/simplify', '-i', output_file, '-o', out_pth2, '-m', '-c', '1e-2', '-f', '8000', '-r', '0.2']))

    # if int(name[-3:]) == 0:
    #     out = f'/u/haozhang/files/mul_pac/threestudio/outputs/meshes/bone'
    #     if len(name)<= 6:
    #         attach1, v1, p1 = rigging(inn, name, 'quad', out)
    #     else:
    #         attach1, v1, p1 = rigging(inn, name, 'human', out)
```

```python
import os
import numpy as np

def rotate_obj_files(directory):
    """
    旋转指定目录及其子目录下所有 .obj 文件的顶点坐标。
    
    参数:
    - directory: 搜索 .obj 文件的目录路径。
    """
    # 旋转矩阵，顺时针旋转90度围绕X轴
    rotation_matrix = np.array([[1, 0, 0],
                                [0, 0, 1],
                                [0, -1, 0]])

    # 遍历目录
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.obj'):
                filepath = os.path.join(root, file)
                rotate_obj_file(filepath, rotation_matrix)

def rotate_obj_file(filepath, rotation_matrix):
    """
    旋转一个 .obj 文件的顶点。
    
    参数:
    - filepath: .obj 文件的路径。
    - rotation_matrix: 应用到顶点的旋转矩阵。
    """
    transformed_lines = []
    with open(filepath, 'r') as file:
        for line in file:
            if line.startswith('v '):
                parts = line.split()
                vertex = np.array([float(parts[1]), float(parts[2]), float(parts[3])])
                # 旋转顶点
                rotated_vertex = rotation_matrix.dot(vertex)
                # 创建新的行
                new_line = 'v {:.6f} {:.6f} {:.6f}\n'.format(rotated_vertex[0], rotated_vertex[1], rotated_vertex[2])
                transformed_lines.append(new_line)
            else:
                transformed_lines.append(line)

    # 保存修改后的文件
    with open(filepath, 'w') as file:
        file.writelines(transformed_lines)

    print(f"Rotated {filepath}")

# 使用示例
directory = '/u/haozhang/files/4D data/DeformingThings4D/experiments_new2'
rotate_obj_files(directory)
```

```python
import os
import shutil

def organize_ply_files(base_path):
    # List all files in the given base path
    files = os.listdir(base_path)
    
    # Filter out files that match the pattern 'pred{X}.ply'
    pred_files = [f for f in files if f.startswith('pred') and f.endswith('.ply')]
    
    # Loop through the filtered files
    for file_name in pred_files:
        # Extract the number {X} from the file name
        number = file_name[len('pred'):-len('.ply')]
        
        # Create a new folder named 'camel{X}'
        new_folder_name = f'camel{number}'
        new_folder_path = os.path.join(base_path, new_folder_name)
        os.makedirs(new_folder_path, exist_ok=True)
        
        # Move the 'pred{X}.ply' file into the 'camel{X}' folder
        src_file_path = os.path.join(base_path, file_name)
        dst_file_path = os.path.join(new_folder_path, file_name)
        shutil.move(src_file_path, dst_file_path)

# Example usage:
# Given path where 'pred{X}.ply' files are located
base_path = '/scratch/bbqk/haozhang/files/pynocchio/data/sequence_camel/camel-1'

# Call the function to organize the .ply files
organize_ply_files(base_path)
```

```python
from PIL import Image
import os
import numpy as np

def find_object_center(image_path):
    """
    Find the center of the non-background part of an image.
    Assumes background is transparent.
    """
    with Image.open(image_path) as img:
        # Convert image to numpy array
        img_array = np.array(img)
        # Find indices where the image is not transparent (assuming the last channel is alpha)
        rows, cols = np.where(img_array[:, :, -1] != 0)
        # Calculate the center
        center_x, center_y = cols.mean(), rows.mean()
        return int(center_x), int(center_y)

def create_mask(image, crop_box):
    """
    Create a mask for the cropped area of the image.
    """
    mask = Image.new('L', (image.width, image.height), 0)
    for i in range(int(crop_box[1]), int(crop_box[3])):
        for j in range(int(crop_box[0]), int(crop_box[2])):
            if image.getpixel((j, i))[3] != 0:
                mask.putpixel((j, i), 255)
    return mask.crop(crop_box)

def crop_images_and_create_masks(directory, center_x, center_y):
    """
    Crop all images in the directory to a 1:1 aspect ratio based on a given center
    and create corresponding masks.
    """
    mask_directory = os.path.join(directory, 'masks')
    os.makedirs(mask_directory, exist_ok=True)

    for filename in os.listdir(directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')):
            full_path = os.path.join(directory, filename)
            with Image.open(full_path) as img:
                width, height = img.size
                min_dimension = min(width, height)
                left = max(0, center_x - min_dimension // 2)
                top = max(0, center_y - min_dimension // 2)
                left = min(left, width - min_dimension)
                top = min(top, height - min_dimension)
                right = left + min_dimension
                bottom = top + min_dimension

                # Crop the image
                cropped_img = img.crop((left, top, right, bottom))

                # Create mask for the cropped image
                mask = create_mask(img, (left, top, right, bottom))

                # Save the cropped image
                cropped_img.save(full_path)
                print(f"Cropped and saved '{filename}'")

                # Save the mask
                mask_path = os.path.join(mask_directory, filename)
                mask.save(mask_path)
                print(f"Mask saved to '{mask_path}'")

# 主执行部分
directory_path = "/u/haozhang/files/2D data/human_1"
first_image_path = os.path.join(directory_path, os.listdir(directory_path)[0])
center_x, center_y = find_object_center(first_image_path)
crop_images_and_create_masks(directory_path, center_x, center_y)
```

```python
def list_of_lists_to_edges(list_of_lists):
    edges = set()  # 使用集合来自动处理重复的边
    for i, connected_points in enumerate(list_of_lists):
        for point in connected_points:
            # 确保边的顺序一致，比如总是小的index在前
            edge = tuple(sorted([i, point]))
            edges.add(edge)
    # 将集合转换为数组，并确保每个边都是M*2的形式
    return np.array(list(edges))

# 调用函数
edges = list_of_lists_to_edges(lol)
# 输出结果
print(edges.shape)
```
